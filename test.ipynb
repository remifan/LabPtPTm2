```python
def attr_h5_to_zarr(v):
  # convert incompatible dtype
  if isinstance(v, bool) or isinstance(v, np.bool_):
      v = 1 if v else 0
  if isinstance(v, np.int64):
      v = int(v)
  return v
```

```python
root = zarr.open_group('./data.zarr', mode='w')
opt = {}
```

```python
grp_src = root.create_group('source')

with h5py.File('./local/data/source/src1.h5', mode='r') as src:
    grp = grp_src.create_group('src1')
    for k, v in src.attrs.items():
        grp.attrs[k] = attr_h5_to_zarr(v)
    ds = grp.create_dataset('sym_evench', shape=src['sym_evench'].shape, chunks=True, dtype='c8', **opt)
    ds[:] = src['sym_evench'][:]
    ds = grp.create_dataset('sym_oddch', shape=src['sym_oddch'].shape, chunks=True, dtype='c8', **opt)
    ds[:] = src['sym_oddch'][:]
    ds = grp.create_dataset('wav_evench', shape=src['wav_evench'].shape, chunks=True, dtype='c8', **opt)
    ds[:] = src['wav_evench'][:]
    ds = grp.create_dataset('wav_oddch', shape=src['wav_oddch'].shape, chunks=True, dtype='c8', **opt)
    ds[:] = src['wav_oddch'][:]

with h5py.File('./local/data/source/src2.h5', mode='r') as src:
    grp = grp_src.create_group('src2')
    for k, v in src.attrs.items():
        grp.attrs[k] = attr_h5_to_zarr(v)
    ds = grp.create_dataset('sym_evench', shape=src['sym_evench'].shape, chunks=True, dtype='c8', **opt)
    ds[:] = src['sym_evench'][:]
    ds = grp.create_dataset('sym_oddch', shape=src['sym_oddch'].shape, chunks=True, dtype='c8', **opt)
    ds[:] = src['sym_oddch'][:]
    ds = grp.create_dataset('wav_evench', shape=src['wav_evench'].shape, chunks=True, dtype='c8', **opt)
    ds[:] = src['wav_evench'][:]
    ds = grp.create_dataset('wav_oddch', shape=src['wav_oddch'].shape, chunks=True, dtype='c8', **opt)
    ds[:] = src['wav_oddch'][:]
```

```python
grp_wav = root.create_group('1125km_SSMF')

for path,dirs,files in os.walk('./local/data'):
    for f in fnmatch.filter(files,'*.h5'):
        h5srcname = os.path.abspath(os.path.join(path,f))
#         print(h5srcname)
        
        levels = h5srcname.split(os.path.sep)
        if levels[-2] == 'source':
            continue
        
        tm, lp = levels[-2][-4:], levels[-1].split('.')[0]
      
        print(tm, lp)

        with h5py.File(h5srcname, mode='r') as src:
            recv = src['recv'][:]
            sent = src['sent'][:]
            attrs = dict(src.attrs)
            
        if tm not in grp_wav.group_keys():
            grp_tm = grp_wav.create_group(tm)
        else:
            grp_tm = grp_wav[tm]
            
        grp_lp = grp_tm.create_group(lp)
            
        for k, v in attrs.items():
            # convert incompatible dtype
            grp_lp.attrs[k] = attr_h5_to_zarr(v)
            
        ds = grp_lp.create_dataset('sent', shape=sent.shape, chunks=True, dtype='c8', **opt)
        ds[:] = sent
        ds = grp_lp.create_dataset('recv', shape=recv.shape, chunks=True, dtype='c8', **opt)
        ds[:] = recv
```

```python
grp_wav = root.create_group('supdata')

for path,dirs,files in os.walk('./local/supplementary_data'):
    for f in fnmatch.filter(files,'*.h5'):
        h5srcname = os.path.abspath(os.path.join(path,f))
#         print(h5srcname)
        
        levels = h5srcname.split(os.path.sep)
        if levels[-2] == 'source':
            continue
        
        tm, lp = levels[-2][-4:], levels[-1].split('.')[0]
      
        print(tm, lp)

        with h5py.File(h5srcname, mode='r') as src:
            nfo = src['nfo'][:]
            attrs = dict(src.attrs)
            
        if tm not in grp_wav.group_keys():
            grp_tm = grp_wav.create_group(tm)
        else:
            grp_tm = grp_wav[tm]
            
        grp_lp = grp_tm.create_group(lp)
            
        for k, v in attrs.items():
            # convert incompatible dtype
            grp_lp.attrs[k] = attr_h5_to_zarr(v)
            
        ds = grp_lp.create_dataset('nfo', shape=nfo.shape, chunks=True, dtype='f8', **opt)
        ds[:] = nfo
```
